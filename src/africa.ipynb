```python
from download import download
from bs4 import BeautifulSoup
from requests.compat import urlparse
```

```python
def get_row_key(url):
    html = download('get', url) ## 파싱 동영상 url
    dom = BeautifulSoup(html.text, 'lxml') ## dom화
    metatag = dom.select_one("meta[property='og:image']")['content'] ## key찾는 과정
    rowKey = urlparse(metatag).query
    return rowKey
```

```python
rowKey = get_row_key('http://vod.afreecatv.com/PLAYER/STATION/46201530') ## key 찾아와서
key = rowKey[:-1]+'c' ## url 변경
xml=download('get', 'http://videoimg.afreecatv.com/php/ChatLoad.php', param=key) ## xml request
```

```python
import xml.etree.ElementTree as et
```

```python
xmltree = et.XML(xml.text)
```

```python
chatlist = zip(xmltree.findall('chat/n'), xmltree.findall('chat/m'))
```

```python
for chat in chatlist:
    print(chat[0].text, chat[1].text, sep='\t|')
```
